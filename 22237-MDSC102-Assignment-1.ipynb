```python
import numpy as np
import sympy as sy
import math
from mpmath import nsum, exp, inf
from sympy import integrate, log, exp, oo
import random
import sys
from sympy.abc import a
```

```python
def 2a(r):
    x = sy.Symbol("x")
    return sy.integrate(x**r*sy.exp(-x), (x, 0, oo))
print("EX = ",2a(1))
print("EX^2 = ",2a(2))
print("Variance =  ",2a(2)-2a(1)**2)
print("EX^r = ",2a(5))
```

```python
def q2b(r):
    x = sy.Symbol("x")
    return sy.integrate((x**r)*(1/(sy.pi*(1+x**2))), (x, -oo, oo))
print("EX = ",q2b(1))
print("EX^2 = ",q2b(2))
print("Variance = ",q2b(2)-q2b(1)**2)
print("EX^n = ",q2b(3))
```

```python
def q2c(r):
    x = sy.Symbol("x")
    return sy.integrate((x**r)*x*sy.exp(-x), (x, 0, oo))
print("EX = ",q2c(1))
print("EX^2 = ",q2c(2))
print("Variance = ",q2c(2)-q2c(1)**2)
print("EX^n = ",q2c(3))
```

```python
def q2d(r):
    x = sy.Symbol("x")
    p=0.5
    return nsum(lambda x: (x**r)*p*(1-p)**(x-1),[1,inf])
print("EX = ",q2d(1))
print("EX^2 = ",q2d(2))
print("Variance = ",q2d(2)-q2d(1)**2)
print("EX^n = ",q2d(3))
```

```python
x = sy.Symbol("x")
a=random.uniform(0,sys.maxsize)
print("a = ",a)
print("Integral of f(x) = ",sy.integrate(a**2*x*sy.exp(-a*x),(x,0,oo)))
```

```python
def q4(r):
    x = sy.Symbol("x")
    lamb= 0.3
    return nsum(lambda x: (x**r)*sy.exp(-lamb)*(lamb**x/sy.factorial(x)),[0,inf])
print("EX = ",q4(1))
print("EX^2 = ",q4(2))
print("Variance = ",q4(2)-q4(1)**2)
print("EX^n = ",q4(3))
```

```python
def q6(r): 
    x = sy.Symbol("x")
    alpha=2.0
    beta=5.0
    return sy.integrate((x**r)*(beta*alpha)/(x**(beta+1)), (x, alpha, oo))
print("EX = ",q6(1))
print("EX^2 = ",q6(2))
print("Variance = ",q6(2)-q6(1)**2)
print("EX^n = ",q6(4))
```
