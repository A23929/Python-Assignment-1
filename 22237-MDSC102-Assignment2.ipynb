```python
import numpy as np
import sympy as sy
import math
from mpmath import nsum, exp, inf
from sympy import integrate, log, exp, oo
import matplotlib.pyplot as plt
```

$$f(x)={}^nC_xp^x(1-p)^{n-x}, x=0,1,...,n$$ 

```python
def plot_binom(n,p):
    x=[]
    y=[]
    f=lambda x: ((math.factorial(n))/(math.factorial(x)*math.factorial(n-x))) * p**x * (1-p)**(n-x)
    for i in range(n+1):
        y.append(i)
        x.append(f(i))
    plt.bar(y,x,width=.2,color='k')
    plt.plot(y,x)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f'Binomial distribution with n={n},p={p}')
plot_binom(30,0.2)
```

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} 
  \exp\left( -\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^{\!2}\,\right)$$

```python
def plot_normal(u,s): 
    x=np.linspace(u-4*s,u+4*s,1000)
    f=lambda x: (1/math.sqrt(2*math.pi*s**2))*np.exp(-0.5*((x-u)/s)**2)
    plt.plot(x,f(x))
    plt.title(f'Normal distribution with mean $\mu$={u} and variance $\sigma^2$={s**2}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
plot_normal(3,2)
```

$$f(x)=\lambda e^{-\lambda x}, x\geq0$$

```python
def plot_exp(l):
    x=np.linspace(0,5)
    f=lambda x: l*np.exp(-l*x)
    plt.plot(x,f(x))
    plt.title(f'Exponential distribution with parameter $\lambda$')
    plt.xlabel('x')
    plt.ylabel('f(x)')
plot_exp(0.5)
plot_exp(1)
plot_exp(1.5)
plt.legend([r'$\lambda$ = 0.5',r'$\lambda$ = 1',r'$\lambda$ = 1.5'])
```

$$f(x) = e^{ - \lambda } \frac{\lambda ^x }{{x!}}, \;\;\lambda > 0,\; x=0,1,...$$

```python
def plot_poi(l):
    f=lambda x: np.exp(-l)*(l**x/math.factorial(x))
    x=[i for i in range(21)]
    y=[f(j) for j in x] 
    plt.plot(x,y)
    plt.title(f'Poisson distribution with parameter $\lambda$')
    plt.xlabel('x')
    plt.ylabel('f(x)')
plot_poi(1) 
plot_poi(4) 
plot_poi(10) 
plt.legend([r'$\lambda$ = 1',r'$\lambda$ = 4',r'$\lambda$ = 10'])
```

$$\Gamma(\alpha) = \int_{0}^{\infty} x^{\alpha-1} e^{-x} \,dx, \;\alpha > 0$$

```python
def f(a):
    x=sy.Symbol("x")
    return integrate(x**(a-1)*exp(-x), (x, 0, oo))
def plot_gamma(): 
    b=np.arange(1,7)
    y=[f(i) for i in b]
    plt.plot(b,y)
    plt.title(r'Gamma distribution with parameter $\alpha$')
    plt.xlabel('x')
    plt.ylabel('f(x)')
plot_gamma() 
```

$$f(x;\alpha,\lambda) = x^{\alpha-1} e^{-\lambda x} \frac{\lambda^{\alpha}} {\Gamma(\alpha)}, \;\; x > 0; \; \alpha,\lambda > 0$$

```python
def plot_gammadb(a,l):
    g =  f(a)
    x = np.linspace(0,20,100)
    h = lambda x: (x**(a-1)*np.exp(-l*x)*l**a)/g
    plt.plot(x,h(x))
    plt.title(r'Gamma distribution with parameters $\alpha$ and $\lambda$')
    plt.xlabel('x')
    plt.ylabel('f(x)')
plot_gammadb(1,0.5) 
plot_gammadb(2,0.5) 
plot_gammadb(3,0.5) 
plot_gammadb(9,2) 
plt.legend([r'$\alpha$ = 1, $\lambda$ = 0.5',r'$\alpha$ = 2, $\lambda$ = 0.5',r'$\alpha$ = 3, $\lambda$ = 0.5',r'$\alpha$ = 9, $\lambda$ = 2'])
```
